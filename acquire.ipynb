```python
import pandas as pd
import requests
from io import StringIO

url = 'https://docs.google.com/spreadsheets/d/19F4bcARhVG2auNxOD8FzrZ51FdtB8uXYI0Bj0al7SsI/edit?usp=sharing'

csv_export_url = url.replace('/edit?usp=sharing', '/export?format=csv')

response = requests.get(csv_export_url)
content = response.content.decode('utf-8')

data = pd.read_csv(StringIO(content))
```

```python
### Now we need to save it so we have the file locally or in otherwords we need to cache
# the import so it is locally avaliable and not depended on internet connection. 
data.to_csv('edu_analysis.csv')
```

```python


def aquire_edu_data():
    if os.path.exists('edu_analysis.csv'):
        df = pd.read_csv('edu_analysis.csv')
    else:
        url = 'https://docs.google.com/spreadsheets/d/19F4bcARhVG2auNxOD8FzrZ51FdtB8uXYI0Bj0al7SsI/edit?usp=sharing'

        csv_export_url = url.replace('/edit?usp=sharing', '/export?format=csv')

        response = requests.get(csv_export_url)
        content = response.content.decode('utf-8')

        df = pd.read_csv(StringIO(content))
    
    return df
    
```
